```python
!pip install scikit-fuzzy
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import numpy as np
```

Ejercicio Dos

```python
Temp=ctrl.Antecedent(np.arange(60,110,1),'Temperatura')
Temp['Bajo']=fuzz.trapmf(Temp.universe,[60,60,70,80])
Temp['Alto']=fuzz.trapmf(Temp.universe,[70,80,90,100])
Temp['Super alto']=fuzz.trapmf(Temp.universe,[90,100,110,110])
Temp.view()
```
